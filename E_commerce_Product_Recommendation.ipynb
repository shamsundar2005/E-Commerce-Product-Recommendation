```python
def sort_recommendor(col='rating',sort_type = False):
    """
    A recommendor based on sorting products on the column passed.
    Arguments to be passed:

    col: The Feature to be used for recommendation.
    sort_type: True for Ascending Order
    """
    rated_recommend = df.copy()
    if rated_recommend[col].dtype == 'O':
        col='rating'
    rated_recommend = rated_recommend.sort_values(by=col,ascending = sort_type)
    return rated_recommend[['product','brand','sale_price','rating']].head(10)
```

```python
def sort_recommendor(col='rating',sort_type = False):
    """
    A recommendor based on sorting products on the column passed.
    Arguments to be passed:

    col: The Feature to be used for recommendation.
    sort_type: True for Ascending Order
    """
    rated_recommend = df.copy().loc[df['rating'] >= 3.5]
    if rated_recommend[col].dtype == 'O':
        col='rating'
    rated_recommend = rated_recommend.sort_values(by=col,ascending = sort_type)
    return rated_recommend[['product','brand','sale_price','rating']].head(10)
```

```python
indices = pd.Series(df.index, index=df['product']).drop_duplicates()

def get_recommendations_1(title, cosine_sim=cosine_sim):

    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return df['product'].iloc[movie_indices]
```

```python
rmv_spc = lambda a:a.strip()
get_list = lambda a:list(map(rmv_spc,re.split('& |, |\*|\n', a)))
```

```python
get_list('A & B, C')
```

```python
for col in ['category', 'sub_category', 'type']:
    df2[col] = df2[col].apply(get_list)
```

```python
df2.head()
```

chocolate ice cream not same as chocolate bar

```python
def cleaner(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
for col in ['category', 'sub_category', 'type','brand']:
    df2[col] = df2[col].apply(cleaner)
```

```python
df2.head()
```

```python
def couple(x):
    return ' '.join(x['category']) + ' ' + ' '.join(x['sub_category']) + ' '+x['brand']+' ' +' '.join( x['type'])
df2['soup'] = df2.apply(couple, axis=1)
```

```python
import numpy as np
import pandas as pd
from fuzzywuzzy import process

def get_recommendations_fuzzy_match(word):
    choices = df2['product'].tolist()
    matching_products = process.extractBests(word, choices, score_cutoff=90)

    if matching_products:
        matching_product_names = [product[0] for product in matching_products]
        matching_products_df = df2[df2['product'].isin(matching_product_names)]
        return matching_products_df['product'].tolist()
    else:
        return []

user_input_word = input("Enter a word: ")

matching_products = get_recommendations_fuzzy_match(user_input_word)

if matching_products:
    print(f"Products matching '{user_input_word}':")

    num_products_to_use = min(1, len(matching_products))
    for i in range(num_products_to_use):
        user_input_item = matching_products[i]
        new_rec = get_recommendations_2(user_input_item, cosine_sim2)
        print("\nRecommendations for", user_input_item, ":", new_rec)
else:
    print(f"No products found matching '{user_input_word}'.")
```
